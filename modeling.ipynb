```python
#데이터 불러오기
new_data = pd.read_csv("/Users/yujinkim/Desktop/코멘토_빅데이터구축및분석(21.07.27~)/data_for_model.csv", encoding='utf-8')

new_data = new_data.drop(['Unnamed: 0'], axis=1)
new_data
```

```python
#feature와 label 나누기 

label = new_data['SalePrice']
features = new_data.drop(['SalePrice'],axis=1, inplace = False)

X_train, X_test, Y_train, Y_test = train_test_split(features, label, test_size=0.2, random_state=1)
```

```python
#Linear Regression, Lasso, Ridge 시도
linear = LinearRegression()
lasso = Lasso()
ridge = Ridge()

#학습
linear.fit(X_train, Y_train)
lasso.fit(X_train,Y_train)
ridge.fit(X_train, Y_train)
```

```python
#rmse 이용 평가 함수 생성 
def get_rmse(model):
    pred = model.predict(X_test)
    mse = mean_squared_error(Y_test, pred)
    rmse = np.sqrt(mse)
    print(model.__class__.__name__, 'RMSE:', np.round(rmse,3))
    return rmse
```

```python
def get_rmses(models):
    rmses = []
    for model in models:
        rmse = get_rmse(model)
        rmses.append(rmse)
    return rmses
```

```python
models = [linear, lasso, ridge]
get_rmses(models)
```
